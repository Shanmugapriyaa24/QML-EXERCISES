# OR GATE

```python
import numpy as np
from qiskit import *
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_state_city, plot_state_qsphere,plot_bloch_multivector
%matplotlib inline
```

```python
circ = QuantumCircuit(1)
circ
```

```python
circ.h(0)
circ.draw('mpl')
```

```python
state = Statevector.from_instruction(circ)
print("Statevector of the OR circuit:")
print(state)
```

```python
probabilities = state.probabilities_dict()
print("Measurement probabilities:")
print(probabilities)
```

```python
plot_state_qsphere(state.data)
```

```python
plot_bloch_multivector(state.data)
```

# AND GATE

\begin{align}
\text{Inputs} \quad a, b, c &\quad \text{Output} \\
\hline
0, 0, 0 &\quad 0 \\
0, 0, 1 &\quad 0 \\
0, 1, 0 &\quad 0 \\
0, 1, 1 &\quad 0 \\
1, 0, 0 &\quad 0 \\
1, 0, 1 &\quad 0 \\
1, 1, 0 &\quad 0 \\
1, 1, 1 &\quad 1 \\
\end{align}

```python
circ = QuantumCircuit(3, 1)
circ.draw()
```

```python
# a = 1
circ.x(0)
circ.draw()

#  b = 1
circ.x(1)
circ.draw()
```

```python
circ.ccx(0, 1, 2)
circ.draw('mpl')
```

```python
state = Statevector.from_instruction(circ)
print("Statevector of the AND circuit:")
print(state)
```

```python
probabilities = state.probabilities_dict()
print("Measurement probabilities:")
print(probabilities)
```

```python
plot_state_qsphere(state.data)
```

```python
plot_bloch_multivector(state.data)
```
